```python
def get_1d(array):
    x = array.transpose()[0]
    y = array.transpose()[1]
    return x * 17 + y

def get_2d(array):
    x = np.trunc(array/17)
    y = array-x*17
    return np.stack((x,y)).transpose()
def in_field(array):
    lower_border = 0
    upper_border = 16
    x = array.transpose()[0]
    y = array.transpose()[1]
    index = (x >= lower_border) & (y >= lower_border) & (x <= upper_border) & (y <= upper_border)
    return array[index]

```
