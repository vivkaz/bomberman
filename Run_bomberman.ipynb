```python
def get_1d(array):
    x = array.transpose()[0]
    y = array.transpose()[1]
    return x * 17 + y

def get_2d(array):
    x = np.trunc(array/17)
    y = array-x*17
    return np.stack((x,y)).transpose()
def in_field(array):
    lower_border = 0
    upper_border = 16
    x = array.transpose()[0]
    y = array.transpose()[1]
    index = (x >= lower_border) & (y >= lower_border) & (x <= upper_border) & (y <= upper_border)
    return array[index]

```

```python
test = np.array([[0,1],[-2,10],[-5,4],[18,15],[5,7]])
test = in_field(test)
print(test)
d1 = get_1d(test)
print(d1)
d2 = get_2d(d1)
print(d2)
```
